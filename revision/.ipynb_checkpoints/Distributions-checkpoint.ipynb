```python
%matplotlib inline

import math
import matplotlib.pyplot as plt
```

```python
def binomial(n,k,p):
    return math.comb(n,k) * p ** k * (1-p) **(n-k)


n = 10
p1 = 0.1
p2 = 0.5

ks = [i for i in range(10)]

plt.bar(ks,[binomial(n,k,p1) for k in ks])
plt.title('Binomial with p = 0.1')
plt.xlabel('x')
plt.ylabel('p(x)')
```

```python
plt.bar(ks,[binomial(n,k,p2) for k in ks])
plt.title('Binomial with p = 0.5')
plt.xlabel('x')
plt.ylabel('p(x)')
```

```python
def neg_bin(r,k,p):
    return math.comb(k-1,r-1)*p**r * (1-p)**(k-r)
   
xs = [i for i in range(1,50)]
p3 = 1/9
r = 2
plt.plot(xs,[neg_bin(r,x,p3) for x in xs], 'bo')
plt.title('Negative Binomial with p = 1/9 and r = 2')
plt.xlabel('x')
plt.ylabel('p(x)')
```
